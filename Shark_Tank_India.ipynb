```python
df['Anupam']=df['Anupam'].map({'Y' : 'Yes', 'N' : 'No'})
df['Namita']=df['Namita'].map({'Y': 'Yes','N': 'No'})
df['Aman']=df['Aman'].map({'Y':'Yes','N':'No'})
df['Peyush']=df['Peyush'].map({'Y' :'Yes','N':'No'})
df['Vineeta']=df['Vineeta'].map({'Y':"Yes",'N':"No"})
df['Ghazal']=df['Ghazal'].map({'Y':"Yes",'N':"No"})
df['Ashneer']=df['Ashneer'].map({'Y':"Yes",'N':"No"})
```

Calculating the Valuation of the Company

```python
df.dtypes
```

```python
df.rename(columns={'Equity':'Equity (In Percentage)'}, inplace=True)
```

```python
df['Equity (In Percentage)']=pd.to_numeric(df['Equity (In Percentage)'].str.replace('%',''))/100
```

```python
df
```

```python
df['Valuation']=df['Investment Amount(lakhs)']/ df['Equity (In Percentage)']
```

```python
df
```

```python
df.rename(columns={'Valuation': 'Valuation(lakhs)'}, inplace=True)
```

```python
df['Valuation(lakhs)'].fillna(0)
```

```python
df['Valuation(lakhs)']=df['Valuation(lakhs)'].round(2)
```
